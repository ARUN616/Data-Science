```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_excel('Data_Train.xlsx')
df_test=pd.read_excel('Test_set.xlsx')
```

```python
final_df=pd.concat([df_train,df_test])
```

```python
final_df
```

```python
#Handling the category column
final_df['Date']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
final_df['Month']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
final_df['Year']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
final_df.head(2)
```

```python
final_df.drop('Date_of_Journey',inplace=True,axis=1)
```

```python
final_df.head(2)
```

```python
final_df.drop('Route',inplace=True,axis=1)
```

```python
final_df['Dept_Hour']=final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
final_df['Dept_Min']=final_df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.head(2)
```

```python
final_df.drop('Dep_Time',inplace=True,axis=1)
```

```python
final_df.head(2)
```

```python
final_df['Arrival_Duration']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head(2)
```

```python
final_df.drop('Arrival_Time',inplace=True,axis=1)
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Duration'].apply(lambda x:x.split(':')[0])
final_df['Arrival_Min']=final_df['Arrival_Duration'].apply(lambda x:x.split(':')[1])
final_df.drop('Arrival_Duration',inplace=True,axis=1)
```

```python
final_df.head(2)
```

```python
final_df['Duration_Hour']=final_df['Duration'].apply(lambda x:x.split(' ')[0].split('h')[0])
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
final_df['Total_Stops'] = le.fit_transform(final_df['Total_Stops'])
```

```python
final_df.drop('Duration',inplace=True,axis=1)
```

```python
final_df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create a dummy employee dataset
data = {
    'Employee id': [10, 20, 15, 25, 30],
    'Gender': ['M', 'F', 'F', 'M', 'F'],
    'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice'],
}

# Convert to a Pandas dataframe
df = pd.DataFrame(data)

# Extract categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Initialize OneHotEncoder with sparse_output=True
encoder = OneHotEncoder(sparse_output=True)

# Apply one-hot encoding to the categorical columns
one_hot_encoded_sparse = encoder.fit_transform(df[categorical_columns])

# Display the resulting sparse matrix
print(one_hot_encoded_sparse)
```
