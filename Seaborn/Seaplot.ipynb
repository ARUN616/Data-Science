```python
# Facet grid

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Extracurricular': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Exam_Score': [85, 70, 90, 75, 80, 65, 88, 72]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a facet grid
g = sns.FacetGrid(df, col='Gender', row='Extracurricular')
# Map the exam scores to the grid
g.map(plt.hist, 'Exam_Score')

# Add titles to each subplot
g.set_titles(row_template = '{row_name}', col_template = '{col_name}')


plt.show()
```

```python
# Distribution observation
sns.violinplot(data,x='specialisation',y='mba_p',hue='gender',loc=0,bbox_to_anchor=(1, 0.5))
```
