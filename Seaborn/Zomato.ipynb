### Q10) Find the list of Affordable Restaurants?
The criteria for Affordable Restaurants would be:-
1) Low Price 2) High Rated

First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
restaurants that are 1/4th of 6000.

```python
max_price=df_zomato['PRICE'].max()
one_fourth_price=max_price/4
one_fourth_price
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restauran
df_aff_res = df_zomato[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE']]
df_aff_res=df_aff_res[df_aff_res['PRICE']<=1250]
df_aff_res.sort_values(by='PRICE',inplace=True)
df_aff_res.head(1)
```

```python
# Finding the highest rated list of restaurants
df_high_rat = df_zomato[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE','RATING']]
df_high_rat=df_high_rat[df_high_rat['RATING']>=4.5]
df_high_rat.sort_values(by='PRICE',inplace=True)
df_high_rat.head(1)
```

```python
#Now, we'll merge the df_aff_res with df_high_rat to obtain the intersection i.e the list of Affordable Restaurants !!
df_high_rate=pd.merge(df_aff_res,df_high_rat,how='inner',on=['NAME','REGION'])
df_high_rate.columns
```

```python
df_high_rate = df_high_rate[['NAME', 'PRICE_x', 'CUSINE_CATEGORY_x', 'REGION', 'CUSINE TYPE_x']]
df_high_rate.rename(columns={'NAME':'NAME', 'PRICE_x':'PRICE', 'CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
 'REGION':'REGION', 'CUSINE TYPE_x':'CUSINE TYPE'},inplace=True)
```

```python
## Affordable Restaurants with low price and high rating 
df_high_rate
```

The criteria for most Reliable Restaurants would be:-
1) Low Price 2) High Rated 3) Large No. of Votes

First step will be to find the restaurants with Votes greater than Mean of Votes 

```python
df_mean_votes=df_zomato['VOTES'].mean()
df_mean_votes
```

```python
df_mean_rest = df_zomato[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE', 'VOTES']]
df_mean_rest=df_mean_rest[df_mean_rest['VOTES']>177.265]
df_mean_rest.sort_values(by='VOTES',inplace=True)
df_mean_rest.head(2)
```

- Cost is below 1250
- Rating is above 4.5
- Votes are above 177
